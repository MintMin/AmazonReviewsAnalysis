```python
import google.cloud.bigquery as bq
```

```python
client = bq.Client.from_service_account_json(r"C:\Users\HYZ7C\Desktop\Amazon Kaggle\Amazon Fine Food Reviews\service account\composite-cable-267902-437712ebc1d4.json")
```

```python
df = client.query("""
SELECT
    Id,
    ProductId,
    UserId,
    ProfileName,
    HelpfulnessNumerator,
    HelpfulnessDenominator,
    Score, 
    DATE(TIMESTAMP_SECONDS(Time)) as Date,
    Summary,
    Text,
FROM `composite-cable-267902.fine_food.Reviews`
    """).to_dataframe()
```

```python
df.Date = df.Date.apply(lambda x: str(x))
df['Year'] = df.Date
df['Year'] = df['Year'].apply(lambda x: x[:-6])
df.sort_values(by='Year')
```

```python
df = df[df['Year'] > '2006']
```

```python
cleanReviews = df.copy()
```

```python
columns = ['HelpfulnessNumerator', 'HelpfulnessDenominator', 'Score', 'Count'] # Where Count is the number of duplciates
for i in columns:
    print(i)
    if(i == 'Count'):
        cleanReviews[i] = cleanReviews.groupby(['UserId','ProductId','Summary','Text'])['Score'].transform('count')
    else:
        cleanReviews[i] = cleanReviews.groupby(['UserId','ProductId','Summary','Text'])[i].transform('mean')

# Now, remove duplicates, and keep first entry
cleanReviews.drop_duplicates(subset = ['UserId', 'ProductId', 'Summary', 'Text'], 
                     keep = 'first', inplace = True) 
print('removed',df.shape[0] - cleanReviews.shape[0], 'entries, leaving us with a total of:', cleanReviews.shape[0])
```

```python
# If we remove duplicates of user and reviews, that should account for almost all of them
for i in columns:
    print(i)
    if(i == 'Count'):
        cleanReviews[i] = cleanReviews.groupby(['UserId','ProductId','Summary','Text'])['Score'].transform('count')
    else:
        cleanReviews[i] = cleanReviews.groupby(['UserId','ProductId','Summary','Text'])[i].transform('mean')

# Now, remove duplicates, and keep first entry
cleanReviews.drop_duplicates(subset = ['UserId', 'Year','Summary', 'Text'], 
                     keep = 'first', inplace = True) 
print('removed',df.shape[0] - cleanReviews.shape[0], 'entries, leaving us with a total of:', cleanReviews.shape[0])
```

```python
import re
def cleanhtml(raw_html):
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', raw_html)
    return cleantext
cleanReviews.Text = cleanReviews.Text.apply(cleanhtml)
```

```python
xTrain = df[['Text','Score']]
```

```python
xTrain.drop_duplicates(subset ="Text", 
                     keep = False, inplace = True) 
```

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python
max_value = df.Score.max()
min_value = df.Score.min()
xTrain.Score = (xTrain.Score - min_value) / (max_value - min_value)
```

```python
import re
import string
def remove_punct(text):
    text_nopunct = ''
    text_nopunct = re.sub('['+string.punctuation+']', '', text)
    return text_nopunct
xTrain['Text'] = xTrain['Text'].apply(lambda x: remove_punct(x))
```

```python
xTrain['Score'] = xTrain['Score'].astype(str)
xTrain['CSV'] = xTrain[['Text', 'Score']].agg(','.join, axis=1)
```
