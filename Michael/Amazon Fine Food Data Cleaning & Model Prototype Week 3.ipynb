```python
import google.cloud.bigquery as bq
```

```python
client = bq.Client.from_service_account_json(r"C:\Users\HYZ7C\Desktop\Amazon Kaggle\Amazon Fine Food Reviews\service account\composite-cable-267902-437712ebc1d4.json")
```

```python
df = client.query("""
SELECT
    Id,
    ProductId,
    UserId,
    ProfileName,
    HelpfulnessNumerator,
    HelpfulnessDenominator,
    Score, 
    DATE(TIMESTAMP_SECONDS(Time)) as Date,
    Summary,
    Text,
FROM `composite-cable-267902.fine_food.Reviews`
    """).to_dataframe()
```

```python
df.Date = df.Date.apply(lambda x: str(x))
df['Year'] = df.Date
df['Year'] = df['Year'].apply(lambda x: x[:-6])
df.sort_values(by='Year')
```

```python
df = df[df['Year'] > '2006']
```

```python
df.head()
```

```python
df['Text'] = df['Text'].str.replace(r'<br />', '')
```

```python
df.Text.head()
```

```python
df[df['UserId'] == 'A3PJZ8TU8FDQ1K']
```

```python
y = df.Score
X = df.Text
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
xTrain.to_csv(r'C:\Users\HYZ7C\Desktop\Amazon Kaggle\Amazon Fine Food Reviews\xTrain.csv',index=False)
xTest.to_csv(r'C:\Users\HYZ7C\Desktop\Amazon Kaggle\Amazon Fine Food Reviews\xTest.csv',index=False)
yTrain.to_csv(r'C:\Users\HYZ7C\Desktop\Amazon Kaggle\Amazon Fine Food Reviews\yTrain.csv',index=False)
yTest.to_csv(r'C:\Users\HYZ7C\Desktop\Amazon Kaggle\Amazon Fine Food Reviews\yTest.csv',index=False)
```
